```python
df['time'] = pd.DatetimeIndex(pd.date_range('2016-11-01 00:01', periods=len(df),  freq='min'))
df = df.set_index('time')
```

```python
df['month'] = df.index.month
df['day'] = df.index.day
df['weekday'] = df.index.day_name()
df['hour'] = df.index.hour
df['minute'] = df.index.minute
df.tail(2)
```

```python
interested = ['Barn_r', 'Barn_p']
df1 = df[interested]
df1['time'] = pd.DatetimeIndex(pd.date_range('2016-11-01 00:01', periods=len(df),  freq='min'))
df1 = df1.set_index('time')
df1['day'] = df1.index.day
df1['hour'] = df1.index.hour
df_1l = df1[df1['hour'].isin(low)]
df_1l['fare'] = 107.0
df_1m = df1[df1['hour'].isin(mid)]
df_1m['fare'] = 153.0
df_1h = df1[df1['hour'].isin(high)]
df_1h['fare'] = 188.8
df1 = pd.concat([df_1l,df_1m,df_1h])
df1 = df1.sort_index()
```

```python
df1['Barn_r_fare'] = df1['Barn_r'] * df1['fare'] / 60
df1['Barn_p_fare'] = df1['Barn_p'] * df1['fare'] / 60
```

```python
df1_1 = df1.loc[df1['day']==1]
#하루요금량
print(sum(df1_1['Barn_r_fare']))
print(sum(df1_1['Barn_p_fare']))
```
