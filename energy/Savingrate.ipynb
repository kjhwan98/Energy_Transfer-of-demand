```python
month_usage = 0 # 11월 사용량 초기화
all_result = 0 # 절감률 초기화

# 11월 1일부터 30일까지 반복
for i in range(1,31):
    df_d = df.loc[df['day']==i]

    # 최대부하, 중간부하, 경부하 정의
    df1_d = df_d[df_d['hour'].isin(high)]
    df2_d = df_d[df_d['hour'].isin(mid)]
    df3_d = df_d[df_d['hour'].isin(low)]
    
    # 최대부하, 중간부하, 경부하 전력사용량 총합
    total=0
    for j in cols:
        a = sum(df1_d[j])
        total+=a
    total2=0
    for j in cols:
        a = sum(df2_d[j])
        total2+=a
    total3=0
    for j in cols:
        a = sum(df3_d[j])
        total3+=a
    
    # 기기별 수요이전 하지 않는 전력사용량    
    Barn0_d1 = Barn0.loc[Barn0['day']==i]
    Dish0_d1 = Dish0.loc[Dish0['day']==i]
    Fridge0_d1 = Fridge0.loc[Fridge0['day']==i]
    Furnace0_d1 = Furnace0.loc[Furnace0['day']==i]
    Garagedoor0_d1 = Garagedoor0.loc[Garagedoor0['day']==i]
    office0_d1 = office0.loc[office0['day']==i]
    Kitchen0_d1 = Kitchen0.loc[Kitchen0['day']==i]
    Living0_d1 = Living0.loc[Living0['day']==i]
    Microwave0_d1 = Microwave0.loc[Microwave0['day']==i]
    Solar0_d1 = Solar0.loc[Solar0['day']==i]
    Well0_d1 = Well0.loc[Well0['day']==i]
    Wine0_d1 = Wine0.loc[Wine0['day']==i]
    
    # 기기별 수요이전 하지 않는 전력사용량 총합
    all0_d1 = sum(Barn0_d1['Barn_p']) + sum(Dish0_d1['Dishwasher_p']) + sum(Fridge0_d1['Fridge_p']) 
    + sum(Furnace0_d1['Furnace_p']) + sum(Garagedoor0_d1['Garage door_p']) + sum(office0_d1['Home office_p'])
    + sum(Kitchen0_d1['Kitchen_p']) + sum(Living0_d1['Living room_p']) + sum(Microwave0_d1['Microwave_p']) 
    + sum(Solar0_d1['Solar_p']) + sum(Well0_d1['Well_p']) + sum(Wine0_d1['Wine cellar_p'])
    
    # 기기별 수요이전 하는 전력사용량
    Barn1_d1 = Barn1.loc[Barn1['day']==i]
    Dish1_d1 = Dish1.loc[Dish1['day']==i]
    Fridge1_d1 = Fridge1.loc[Fridge1['day']==i]
    Furnace1_d1 = Furnace1.loc[Furnace1['day']==i]
    Garagedoor1_d1 = Garagedoor1.loc[Garagedoor1['day']==i]
    office1_d1 = office1.loc[office1['day']==i]
    Kitchen1_d1 = Kitchen1.loc[Kitchen1['day']==i]
    Living1_d1 = Living1.loc[Living1['day']==i]
    Microwave1_d1 = Microwave1.loc[Microwave1['day']==i]
    Solar1_d1 = Solar1.loc[Solar1['day']==i]
    Well1_d1 = Well1.loc[Well1['day']==i]
    Wine1_d1 = Wine1.loc[Wine1['day']==i]
    
    # 기기별 수요이전 하는 전력사용량 총합
    all1_d1 = sum(Barn1_d1['Barn_p']) + sum(Dish1_d1['Dishwasher_p']) + sum(Fridge1_d1['Fridge_p'])
    + sum(Furnace1_d1['Furnace_p']) + sum(Garagedoor1_d1['Garage door_p']) + sum(office1_d1['Home office_p']) 
    + sum(Kitchen1_d1['Kitchen_p']) + sum(Living1_d1['Living room_p']) + sum(Microwave1_d1['Microwave_p']) 
    + sum(Solar1_d1['Solar_p']) + sum(Well1_d1['Well_p']) + sum(Wine1_d1['Wine cellar_p'])

    all_total = total+total2+total3
    print(f'11월 {i}일의 전력 사용량: {all_total}')
    
    # 11월 총 사용량 합계
    month_usage +=all_total
    
    a = (total*188.8 + total2*153.0 + total3*107.0)/60
    print(f'11월 {i}일의 수요이전 수행 전 전력요금: {a}')
    b = (((all0_d1*188.8)+(all1_d1*107.0))+ total2*153.0 + total3*107.0)/60
    print(f'11월 {i}일의 수요이전 수행 후 전력요금: {b}')
    result = (a-b)/a*100
    print(f'11월 {i}일의 요금 절감률: {result}')
    print('------------------------------------------')
    
    all_result += result

print(f'11월의 총 전력 사용량: {month_usage}')

result_avg = all_result/30
print(f'11월 절감률 평균: {result_avg}')
    
```
