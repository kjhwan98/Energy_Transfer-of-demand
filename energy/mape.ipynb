```python
#1분단위의 데이터
df = pd.read_csv("HomeC_11f.csv")
df
```

```python
#1일 단위의 사용량과 요금
df1 = df.groupby("day")['real','predict','r_fare','p_fare'].sum()
df1
```

```python
#사용량 예측 모델과 요금 모델 정확도 계산

#사용량 예측 모델
df1['mape1'] =np.abs((df1['real'] - df1['predict'])/df1['real']*100)
#요금 예측모델
df1['mape2'] =np.abs((df1['r_fare'] - df1['p_fare'])/df1['r_fare']*100)
df1
```

```python
#사용량 예측 모델정확도
np.mean(df1['mape1'])
```

```python
#요금 모델 정확도
np.mean(df1['mape2'])
```
